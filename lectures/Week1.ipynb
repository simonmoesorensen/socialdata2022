```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc = {'figure.figsize':(15,8)})
```

```python
group = df.groupby('Category').count()['IncidntNum']
```

```python
print(f'Most frequent category of crime: {group.idxmax()}')
```

```python
print(f'Least frequent category of crime: {group.idxmin()}')
```

```python
sorted_ = group.sort_values(ascending=False)
```

```python
sns.barplot(x=sorted_.index, y=sorted_)
plt.xticks(rotation=85)
plt.title('Number of incidents for each category')
plt.ylabel('Frequency')
plt.tight_layout()
```

The plot above already gives us a bit of information but does not tell us anything about how crimes changed over the years. Let's have a look at it!

> *Exercise 3*: Temporal patterns.
> * To start off easily, let's count the number of crimes per year:
>     * What is the year with most crimes?

> * Now, let's change our bar-plot above by replacing the count with the *average number of crimes per year by category* and by adding its *standard deviation*.
> * Finally, Police chief Suneman is interested in the temporal development of only a subset of categories, the so-called focus crimes. Those categories are listed below (for convenient copy-paste action). Now create bar-charts displaying the year-by-year development of each of these categories across the years 2003-2017.
>

```python
df.Date = pd.to_datetime(df.Date)
```

```python
crimes_pr_year = df.groupby(df.Date.dt.year).count()['IncidntNum']
```

```python
print(f"Year with most crimes: {crimes_pr_year.idxmax()} (count: {crimes_pr_year.max()})")
print(f"Year with least crimes: {crimes_pr_year.idxmin()} (count: {crimes_pr_year.min()})")
```

As we have 2018 with the least number of crimes, we investigate further. By reading the **documentation** of the dataset at https://data.sfgov.org/Public-Safety/Police-Department-Incident-Reports-Historical-2003/tmnf-yvry we see that the data is **incomplete** as it stops in 2018 May, thus we do not have a full year of data points (the data was moved to a new dataset due to unreliability issues). Therefore, to get the correct result, we should exclude 2018.

```python
print(f"Year with least crimes: {crimes_pr_year.drop(2018).idxmin()} (count: {crimes_pr_year.drop(2018).min()})")
```

```python
print(f"Average number of crimes per year: {crimes_pr_year.mean()}")
print(f"Average number of crimes per year (excluding 2018): {crimes_pr_year.drop(2018).mean()}")
```

```python
crimes_pr_year_pr_cat = df.groupby([df.Date.dt.year, df.Category], as_index=False).count()
```

```python
plot_order = crimes_pr_year_pr_cat.groupby('Category')['IncidntNum'].mean().sort_values(ascending=False).index.values
```

```python
sns.color_palette()[1]
```

```python
sns.color_palette()
```

```python
import matplotlib

g = sns.barplot(x=crimes_pr_year_pr_cat.Category, y=crimes_pr_year_pr_cat.IncidntNum, ci='sd', order=plot_order)
plt.xticks(rotation=85)
plt.title('Average number of incidents for each category pr year')
plt.ylabel('Mean')

plt.legend(handles=[matplotlib.patches.Patch(color=sns.color_palette()[0], label='Mean'), matplotlib.patches.Patch(color=(0.2, 0.2, 0.2), label='Error')])

plt.tight_layout()
```

```python
focuscrimes = {'WEAPON LAWS', 'PROSTITUTION', 'DRIVING UNDER THE INFLUENCE', 'ROBBERY', 'BURGLARY', 'ASSAULT',
               'DRUNKENNESS', 'DRUG/NARCOTIC', 'TRESPASS', 'LARCENY/THEFT', 'VANDALISM', 'VEHICLE THEFT',
               'STOLEN PROPERTY', 'DISORDERLY CONDUCT'}
```

```python
df_focus = df[df.Category.isin(focuscrimes) & (2003 <= df.Date.dt.year) & (df.Date.dt.year <= 2017)]
```

```python
df_plot = df_focus.groupby([df_focus.Category, df_focus.Date.dt.year]).count().IncidntNum.reset_index()
```

```python
g = sns.FacetGrid(df_plot, col='Category', height=4, aspect=1.5, col_wrap=2, sharey=False)
g.map(sns.barplot, "Date", "IncidntNum")
g.set_ylabels('Frequency')
g.set_titles('{col_name}')
g.set_xticklabels(rotation=85)
g.fig.subplots_adjust(top=.95)
g.fig.suptitle('Development of focus crimes from 2003 to 2017')
```
